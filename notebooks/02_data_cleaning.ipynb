```python
# 02 - Data Cleaning & Preparation

import pandas as pd
import numpy as np

# Load raw data
df_claims = pd.read_csv('../data/raw/inpatient_claim.csv')
df_bene = pd.read_csv('../data/raw/Bene_File.csv')

# Step 1: Convert Dates from float to datetime
date_cols_claims = ['CLM_FROM_DT', 'CLM_THRU_DT', 'CLM_ADMSN_DT', 'NCH_BENE_DSCHRG_DT']
for col in date_cols_claims:
    df_claims[col] = pd.to_datetime(df_claims[col], errors='coerce', format='%Y%m%d')

# Step 2: Handle Nulls
df_claims.fillna({'CLM_PMT_AMT': 0, 'CLM_PASS_THRU_PER_DIEM_AMT': 0}, inplace=True)
df_bene.fillna({'BENE_SEX_IDENT_CD': 0, 'BENE_RACE_CD': 0}, inplace=True)

# Step 3: Drop Unnecessary Columns (if any obvious ones)

# Step 4: Save Cleaned Output
df_claims.to_csv('../data/cleaned/Inpatient_Claim.csv', index=False)
df_bene.to_csv('../data/cleaned/bene_file.csv', index=False)

print("✅ Data cleaned and saved.")
```
